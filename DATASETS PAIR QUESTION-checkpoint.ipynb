```python
import nltk
from nltk.corpus import stopwords
from string  import punctuation
from nltk.tokenize import  word_tokenize
nltk.download('stopwords')
nltk.download('punkt_tab')
pun = list(punctuation)
stop = stopwords.words('english')
from nltk.stem import SnowballStemmer
sw = SnowballStemmer('english')
```

```python
def transform(text):
    text = text.lower()
    token = nltk.word_tokenize(text)
    y = []
    for i in token:
        if i.isalnum():
            y.append(i)

    text = y[:]
    y.clear()

    for i in text:
        if  i not in pun and i not in stop:
            y.append(i)
    text = y[:]
    y.clear()

    for  i in text:
        stem = sw.stem(i)
        y.append(stem)

    return ' '.join(y)

```

```python
df1.head(1)
```

```python
df1['question1'] = df1['question1'].apply(transform)
```

```python
df1['question2'] = df1['question2'].apply(transform)
```

```python
df1
```

```python
x = df1[['question1','question2']]
y = df1['is_duplicate']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
x_train_bow = cv.fit_transform(x_train['question1']+''+x_train['question2']).toarray()
x_test_bow = cv.transform(x_test['question1']+''+x_test['question2']).toarray()
```

```python
from sklearn.ensemble  import  RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train_bow,y_train)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix
```

```python
y_pred = rf.predict(x_test_bow)
```

```python
accuracy_score(y_pred,y_test)*100
```
